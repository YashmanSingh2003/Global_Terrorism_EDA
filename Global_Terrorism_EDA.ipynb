#### Chart 5

#### which form of attack has given the highest number of kills and casulaties.

```python
data = df1[df1['kills'].notnull()]
data2 = df1[df1['wounded'].notnull()]
attack_types = data['attacktype'].value_counts().index

num_kills = []
num_wounded = []

for attack_type in attack_types:
    num_kills.append(data[data['attacktype'] == attack_type]['kills'].sum())
    num_wounded.append(data2[data2['attacktype'] == attack_type]['wounded'].sum())

# Plotting the data using a bar chart
bar_width = 0.35
index = range(len(attack_types))

plt.figure(figsize=(12, 8))

# Plot bars for kills
plt.bar(index, num_kills, bar_width, label='Kills')

# Plot bars for wounded, with some spacing between the sets of bars
plt.bar([i + bar_width for i in index], num_wounded, bar_width, label='Wounded')

# Adjusting x-ticks to the center of the bars
plt.xticks([i + bar_width / 2 for i in index], attack_types, rotation=90)

plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.title('Kills and Wounded by Attack Type')
plt.legend()

plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

This graph is most suitable because it effectively compares and visualizes trends in kills and wounded across different attack types with clarity and detail using lines and scatter points.

#### Chart 9

#### Which terror group has the highest number of succes rate

```python
df1['output'] = df1['kills'] + df1['wounded']
```

```python
import matplotlib.pyplot as plt

# Initialize an empty dictionary to store success rates
dict1 = dict()

# Get the top 200 terror groups by occurrence
data = df1['Terror group'].value_counts().head(200)

# Calculate success rate for each terror group
for i in range(len(data)):
    # Get the terror group name
    a = data.index[i]
    # Calculate the success rate as the sum of 'output' where 'Terror group' is 'a' divided by the count of 'Terror group' 'a'
    b = df1['output'][df1['Terror group'] == a].sum()
    dict1[a] = round(b / (data.values[i]), 2)

# Sort the dictionary by success rate in descending order
sorted_dict = dict(sorted(dict1.items(), key=lambda item: item[1], reverse=True))

# Get the top 10 terror groups with highest success rates
x = list(sorted_dict.keys())[:10]
y = list(sorted_dict.values())[:10]

# Create a bar plot
plt.bar(x, y)

# Add labels to each bar with the success rate value
for i, value in enumerate(y):
    plt.text(i, value, str(value), ha='center', va='bottom')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Label x-axis as 'Terror groups', y-axis as 'Success Rate', and set title
plt.xlabel('Terror groups')
plt.ylabel('Success Rate')
plt.title('Most Dangerous Terror Groups')

# Display the plot
plt.show()
```

##### 1. Why did you pick the specific chart?
This bar chart is effective for highlighting the most dangerous terror groups by displaying their success rates, allowing for a clear comparison of their relative impact based on the frequency of attacks and their successful outcomes.

##### 2. What are the insight found from the chart?
More than 90% attacks are done by Al-Qaida.
